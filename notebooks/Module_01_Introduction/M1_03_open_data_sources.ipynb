## ğŸ“– Part 1: What is Open Data?

### Definition

**Open data** is data that is:
- âœ… **Freely available** - Anyone can access it
- âœ… **Machine-readable** - Structured format (CSV, JSON, XML)
- âœ… **Licensed for reuse** - Legal to use, modify, and share
- âœ… **Non-discriminatory** - Available to everyone
- âœ… **No registration required** (in many cases)

### Why Open Data Matters

**For Learning & Research:**
- ğŸ“ Practice with real-world data
- ğŸ“Š Build portfolio projects
- ğŸ”¬ Conduct research without cost barriers

**For Society:**
- ğŸ›ï¸ Government transparency
- ğŸ’¡ Innovation and entrepreneurship
- ğŸ‘¥ Citizen engagement
- ğŸ“ˆ Data-driven decision making

**For This Project:**
- ğŸš´ Real bike-sharing data from cities
- ğŸŒ¦ï¸ Real weather data from meteorological institutes
- ğŸ’° **Free** - no API keys or payments required (mostly!)
- ğŸ”„ **Reproducible** - anyone can run the same analysis

---

### Open Data Principles (The 5 Stars â­)

1. â­ **Available online** with open license
2. â­â­ **Structured data** (e.g., Excel instead of PDF)
3. â­â­â­ **Non-proprietary format** (e.g., CSV instead of Excel)
4. â­â­â­â­ **Uses URIs** to identify things
5. â­â­â­â­â­ **Linked data** - connected to other data

**Goal**: We aim for at least 3-star data in this project!

### Legal Considerations

**Always check:**
- ğŸ“œ **License** - Can you use it for education? Commercial use?

### ğŸ§ª Demo: Fetching Bike Data

Let's try fetching real bike-sharing data!

````python
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# DEMO: FETCH BIKE DATA FROM CITYBIKES API
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

print("ğŸš´ Fetching bike-sharing data from CityBikes API...\n")

# CityBikes API endpoint for Amsterdam
url = "http://api.citybik.es/v2/networks/biketep"

try:
    # Make API request
    response = requests.get(url, timeout=10)
    response.raise_for_status()  # Raise error for bad status codes
    
    # Parse JSON response
    data = response.json()
    
    # Extract basic info
    network_info = data['network']
    print(f"âœ… Successfully fetched data!")
    print(f"\nğŸ“Š Network Information:")
    print(f"   Name: {network_info['name']}")
    print(f"   City: {network_info['location']['city']}")
    print(f"   Country: {network_info['location']['country']}")
    print(f"   Total Stations: {len(network_info['stations'])}")
    
    # Look at first few stations
    print(f"\nğŸ” Sample Stations (first 3):\n")
    for station in network_info['stations'][:3]:
        print(f"   Station: {station['name']}")
        print(f"      Location: ({station['latitude']:.4f}, {station['longitude']:.4f})")
        print(f"      Bikes available: {station['free_bikes']}")
        print(f"      Docks available: {station['empty_slots']}")
        print(f"      Timestamp: {station['timestamp']}")
        print()
    
    print("=" * 60)
    print("ğŸ‰ API demo successful! We can fetch real-time bike data!")
    print("=" * 60)
    
except requests.exceptions.RequestException as e:
    print(f"âŒ Error fetching data: {e}")
    print("   Don't worry! We'll work with sample data instead.")
````

### ğŸ§ª Demo: Fetching Weather Data

Let's fetch real weather data!

````python
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# DEMO: FETCH WEATHER DATA FROM OPEN-METEO
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

print("ğŸŒ¦ï¸ Fetching weather data from Open-Meteo API...\n")

# Amsterdam coordinates
latitude = 52.3676
longitude = 4.9041

# Open-Meteo API endpoint (current weather)
url = f"https://api.open-meteo.com/v1/forecast"
params = {
    'latitude': latitude,
    'longitude': longitude,
    'current_weather': True,
    'timezone': 'Europe/Amsterdam'
}

try:
    # Make API request
    response = requests.get(url, params=params, timeout=10)
    response.raise_for_status()
    
    # Parse JSON response
    weather_data = response.json()
    current = weather_data['current_weather']
    
    print(f"âœ… Successfully fetched weather data!")
    print(f"\nğŸŒ¡ï¸ Current Weather in Amsterdam:")
    print(f"   Temperature: {current['temperature']}Â°C")
    print(f"   Wind Speed: {current['windspeed']} km/h")
    print(f"   Wind Direction: {current['winddirection']}Â°")
    print(f"   Time: {current['time']}")
    
    # Weather code interpretation
    weather_codes = {
        0: "Clear sky â˜€ï¸",
        1: "Mainly clear ğŸŒ¤ï¸",
        2: "Partly cloudy â›…",
        3: "Overcast â˜ï¸",
        45: "Foggy ğŸŒ«ï¸",
        48: "Foggy ğŸŒ«ï¸",
        51: "Light drizzle ğŸŒ¦ï¸",
        61: "Light rain ğŸŒ§ï¸",
        80: "Rain showers ğŸŒ§ï¸",
        95: "Thunderstorm â›ˆï¸"
    }
    
    weather_desc = weather_codes.get(current['weathercode'], "Unknown")
    print(f"   Conditions: {weather_desc}")
    
    print("\n" + "=" * 60)
    print("ğŸ‰ Weather API demo successful!")
    print("=" * 60)
    
except requests.exceptions.RequestException as e:
    print(f"âŒ Error fetching data: {e}")
    print("   Don't worry! We'll work with sample data instead.")
````

### ğŸ“Š Weather Data for Time Series

For our prediction model, we'll need **historical** weather data:

````python
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# DEMO: FETCH HISTORICAL WEATHER DATA
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

print("ğŸ“… Fetching historical weather data (past week)...\n")

from datetime import datetime, timedelta

# Date range (past 7 days)
end_date = datetime.now()
start_date = end_date - timedelta(days=7)

# Open-Meteo API endpoint (historical)
url = "https://api.open-meteo.com/v1/forecast"
params = {
    'latitude': latitude,
    'longitude': longitude,
    'start_date': start_date.strftime('%Y-%m-%d'),
    'end_date': end_date.strftime('%Y-%m-%d'),
    'hourly': 'temperature_2m,precipitation,windspeed_10m',
    'timezone': 'Europe/Amsterdam'
}

try:
    response = requests.get(url, params=params, timeout=10)
    response.raise_for_status()
    
    weather_history = response.json()
    hourly = weather_history['hourly']
    
    # Create DataFrame
    weather_df = pd.DataFrame({
        'timestamp': pd.to_datetime(hourly['time']),
        'temperature': hourly['temperature_2m'],
        'precipitation': hourly['precipitation'],
        'windspeed': hourly['windspeed_10m']
    })
    
    print("ğŸ“Š Historical Weather DataFrame:\n")
    print(weather_df.head(10))
    
    print(f"\nğŸ“ˆ Quick Statistics (past week):")
    print(f"   Average temperature: {weather_df['temperature'].mean():.1f}Â°C")
    print(f"   Max temperature: {weather_df['temperature'].max():.1f}Â°C")
    print(f"   Min temperature: {weather_df['temperature'].min():.1f}Â°C")
    print(f"   Total precipitation: {weather_df['precipitation'].sum():.1f} mm")
    print(f"   Average wind speed: {weather_df['windspeed'].mean():.1f} km/h")
    
    print("\nâœ… Historical weather data successfully retrieved!")
    
except Exception as e:
    print(f"Note: Could not fetch historical data.")
    print("We'll work with pre-downloaded data in later notebooks.")
````

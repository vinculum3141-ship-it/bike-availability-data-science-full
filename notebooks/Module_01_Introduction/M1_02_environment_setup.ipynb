Run the cell below to check essential packages:

````python
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# 1. VERIFY CORE PACKAGES
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

print("Checking core data science packages...\n")

packages_to_check = {
    'numpy': 'numpy',
    'pandas': 'pandas',
    'matplotlib': 'matplotlib',
    'seaborn': 'seaborn',
    'scikit-learn': 'sklearn',
    'requests': 'requests',
}

import importlib

all_installed = True
for package_name, import_name in packages_to_check.items():
    try:
        module = importlib.import_module(import_name)
        version = getattr(module, '__version__', 'unknown')
        print(f"âœ… {package_name:20s} version {version}")
    except ImportError:
        print(f"âŒ {package_name:20s} NOT INSTALLED")
        all_installed = False

print("\n" + "=" * 50)
if all_installed:
    print("ğŸ‰ All core packages installed successfully!")
else:
    print("âš ï¸  Some packages are missing. Please install them:")
    print("   pip install -r requirements.txt")
print("=" * 50)
````

````python
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# 3. TEST PANDAS
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

import pandas as pd

# Create a simple DataFrame
test_data = {
    'station': ['Station A', 'Station B', 'Station C'],
    'bikes_available': [5, 0, 12],
    'docks_available': [10, 15, 3]
}

df = pd.DataFrame(test_data)

print("Testing Pandas...")
print("\nSample DataFrame:")
print(df)
print(f"\nTotal bikes: {df['bikes_available'].sum()}")
print("âœ… Pandas is working correctly!")
````

### Test 3: Matplotlib Plotting

````python
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# 4. TEST MATPLOTLIB
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

import matplotlib.pyplot as plt

# Configure matplotlib for notebook display
%matplotlib inline

# Create a simple plot
x = [1, 2, 3, 4, 5]
y = [2, 4, 5, 4, 6]

plt.figure(figsize=(8, 4))
plt.plot(x, y, marker='o', linewidth=2, markersize=8)
plt.title('Test Plot: Matplotlib is Working!')
plt.xlabel('Time')
plt.ylabel('Bike Count')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print("âœ… Matplotlib is working correctly!")
````

### Test 4: Seaborn Styling

````python
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# 5. TEST SEABORN
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

import seaborn as sns

# Set seaborn style
sns.set_style('whitegrid')

# Create a sample plot with seaborn styling
data = pd.DataFrame({
    'day': ['Mon', 'Tue', 'Wed', 'Thu', 'Fri'],
    'rides': [120, 140, 135, 160, 180]
})

plt.figure(figsize=(8, 4))
sns.barplot(data=data, x='day', y='rides', palette='viridis')
plt.title('Test Plot: Seaborn Styling')
plt.ylabel('Number of Rides')
plt.tight_layout()
plt.show()

print("âœ… Seaborn is working correctly!")
````

## âš™ï¸ Part 4: Configure Environment Settings

Let's set up some standard configurations that we'll use throughout the course.

````python
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# 6. STANDARD CONFIGURATION
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

import warnings
import sys
import os

print("Applying standard configurations...\n")

# Pandas display options
pd.set_option('display.max_columns', None)  # Show all columns
pd.set_option('display.max_rows', 100)      # Show up to 100 rows
pd.set_option('display.precision', 2)       # 2 decimal places
pd.set_option('display.float_format', '{:.2f}'.format)

print("âœ… Pandas display options configured")

# Matplotlib style
plt.style.use('seaborn-v0_8-darkgrid')
print("âœ… Matplotlib style set to 'seaborn-darkgrid'")

# Suppress warnings (optional - comment out if you want to see warnings)
warnings.filterwarnings('ignore')
print("âœ… Warnings suppressed")

# Set random seed for reproducibility
RANDOM_SEED = 42
np.random.seed(RANDOM_SEED)
print(f"âœ… Random seed set to {RANDOM_SEED}")

# Verify working directory
print(f"\nğŸ“ Current working directory: {os.getcwd()}")

print("\n" + "=" * 50)
print("ğŸ‰ Environment configured successfully!")
print("=" * 50)
````
